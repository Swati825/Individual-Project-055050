* **5.Data Variable Category : Identifying Categorical (Nominal & Ordinal) and Non - Categorical Data**:

- **Index**:
  - **Transaction_ID**: A unique ID assigned to each transaction.
  - **Invoice_Number**: A distinct number used to identify each invoice.

- **Categorical Variables**:
  - **Nominal Variables**: These include attributes like Country, Product, Import_Export, Category, Port, Shipping_Method, Supplier, Customer, and Payment_Terms, which are categorical in nature and lack any inherent order.

- **Import/Export Status (Categorical Variable)**:  
  - This variable indicates whether the transaction pertains to **import** or **export** activities. It plays a key role in differentiating between incoming and outgoing trade flows, enabling a comparison of factors such as transaction values or quantities between imports and exports. By analyzing the breakdown of these transactions, businesses can detect trade imbalances, such as whether they import more than they export, or vice versa.

- **Shipping Method (Categorical Variable)**:  
  - This variable represents the **mode of transport** (e.g., air, sea, or land) used for shipping goods. The shipping method influences factors like delivery speed, cost, and reliability. Different types of goods require specific shipping methods based on factors such as perishability, cost-effectiveness, and travel distance. Understanding how frequently each method is used can help businesses optimize their logistics, identify faster or more affordable routes, and make informed decisions when negotiating with logistics providers.

- **Payment Terms (Categorical Variable)**:  
  - This variable defines the **terms of payment** for each transaction, such as pre-paid, post-paid, or credit-based terms. Payment terms directly impact a company's cash flow and financial management. Longer credit periods may affect liquidity, while advance payments lower financial risk. By analyzing the payment terms for various transactions, businesses can optimize their financial strategies. For example, frequent use of credit terms might require tighter financial controls, while prepaid terms provide immediate cash flow benefits.

- **Non-Categorical Variables**:
  - Variables like **Quantity, Value, Weight**, and **Customs_Code** are continuous and numerical, representing measurable aspects oand financial planning.
